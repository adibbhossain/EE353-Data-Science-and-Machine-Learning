#### **Summarize** Datasets by ZIP code **using SQL**

```python
import sqlite3  #for creating a database
#convert pandas DataFrames to SQLite tables
conn = sqlite3.connect(':memory:')  #create an in-memory SQLite database
dataframe_H.to_sql('housing_projects', conn, index=False, if_exists='replace')  #load housing data into the database
dataframe_RndM.to_sql('restaurant_inspections', conn, index=False, if_exists='replace')  #load restaurant inspection data into the database

#create a cursor object to execute SQL queries
cur = conn.cursor()

#SQL query to summarize housing data by ZIP code
housing_query = '''
SELECT
    "SITE ADDRESS" AS ZIP_CODE,
    COUNT("PROJECT NUMBER") AS Total_Projects,
    SUM("SITE UNITS") AS Total_Units,
    AVG("TDC") AS Avg_TDC,
    MAX("TDC") AS Max_TDC
FROM
    housing_projects
GROUP BY
    "SITE ADDRESS"
ORDER BY
    Total_Projects DESC
LIMIT 15;  -- Limit the results to the first 15 rows
'''

#SQL query to summarize restaurant inspection data by ZIP code
restaurant_query = '''
SELECT
    facility_zip AS ZIP_CODE,
    COUNT(facility_id) AS Total_Restaurants,
    AVG(score) AS Avg_Health_Score,
    MAX(score) AS Max_Health_Score,
    MIN(score) AS Min_Health_Score
FROM
    restaurant_inspections
GROUP BY
    facility_zip
ORDER BY
    Total_Restaurants DESC
LIMIT 15;  -- Limit the results to the first 15 rows
'''

#execute the queries and fetch the results
housing_summary = pd.read_sql_query(housing_query, conn)  #get housing data summary
restaurant_summary = pd.read_sql_query(restaurant_query, conn)  #get restaurant data summary

#display the summarized data
print("Housing Projects Summary by ZIP Code (First 12 Rows):")
print(housing_summary)  #displaying the summarized housing data

print("\nRestaurant Inspections Summary by ZIP Code (First 12 Rows):")
print(restaurant_summary)  #displaying the summarized restaurant data

#closing the connection to the database
conn.close()
```

####**Joining the files** using SQL by ZIP code

#####**Ensuring** that the ZIP codes are in compatible **formats and lengths**

```python
import re #to checking/matching ZIP code pattern
#for the Restaurant and Market Health Inspections dataset, we have already corrected its format and length

#for Housing Project dataset, we will extract ZIP code from 'SITE ADDRESS' column and put them in a new column
def extract_zip_code(address): #function to extract the ZIP code
    match = re.search(r'\b\d{5}\b', address) #regular expression to find the 5-digit ZIP code at the end of the string
    return match.group(0) if match else '00000'  #return '00000' if no ZIP code is found

#apply the function to the 'address' column and store results in a new column 'ZIP_CODE'
#dataframe_H = pd.DataFrame(housing_data)
dataframe_H['ZIP CODE'] = dataframe_H['SITE ADDRESS'].apply(extract_zip_code)

# Display the DataFrame with the new 'zip_code' column
print(dataframe_H)
```

##### **Getting** the **Predictor variable for dataset1** and **Potential variable for dataset2**

```python
#converting DataFrames to SQLite tables
conn = sqlite3.connect(':memory:')  #ceating an in-memory SQLite database
dataframe_H.to_sql('housing_projects', conn, index=False, if_exists='replace')  #loading housing data into the database
dataframe_RndM.to_sql('restaurant_inspections', conn, index=False, if_exists='replace')  #loading restaurant inspection data into the database

#creating a cursor object to execute SQL queries
cur = conn.cursor()

#SQL query to standardize ZIP codes and join the tables
join_query = '''
WITH
standardized_housing AS (
    SELECT
        "ZIP CODE" AS ZIP_CODE,
        SUM("SITE UNITS") AS Total_Units, -- Predictor variable
        AVG("TDC") AS Avg_TDC
    FROM
        housing_projects
    GROUP BY
        "ZIP CODE"
),
standardized_restaurants AS (
    SELECT
        facility_zip AS ZIP_CODE,
        AVG(score) AS Avg_Health_Score, -- Potential predicted variable 1
        COUNT(facility_id) AS Total_Restaurants -- Potential predicted variable 2
    FROM
        restaurant_inspections
    GROUP BY
        facility_zip
)
SELECT
    h.ZIP_CODE,
    h.Total_Units,
    h.Avg_TDC,
    r.Avg_Health_Score,
    r.Total_Restaurants
FROM
    standardized_housing h
CROSS JOIN
    standardized_restaurants r
ON
    h.ZIP_CODE = r.ZIP_CODE
ORDER BY
    h.ZIP_CODE;
'''

#execute the query and fetch the results
joined_data = pd.read_sql_query(join_query, conn)  #getting joined data

#displaying the joined data
print("Joined Data by ZIP Code:")
print(joined_data)  #priint the joined data

#close the connection to the database
conn.close()
```

####**Formulating** and **Testing** the hypothesis

#####**Formulating a reasonable alternative hypothesis**

###### **Given the datasets, a reasonable alternative hypothesis (Hâ‚) could be:**

**Alternative Hypothesis (Hâ‚):** There is a significant relationship between the total number of housing units (Total_Units) and the average health inspection score (Avg_Health_Score) in a ZIP code.

#####**Formulating a NULL hypothesis**

######**Null Hypothesis (Hâ‚€):** There is no significant relationship between the total number of housing units (Total_Units) and the average health inspection score (Avg_Health_Score) in a ZIP code.

#####**Selecting an Appropriate Test and Significance Level**

######**Test Selection:** Linear Regression is appropriate here because it will help us understand how Total_Units affects Avg_Health_Score.
**Significance Level:** Use a significance level of 0.05.

#####**Performing the test**

```python
#creating an in-memory SQLite database
conn = sqlite3.connect(':memory:')

#convert DataFrames to SQLite tables
dataframe_H.to_sql('housing_projects', conn, index=False, if_exists='replace')
dataframe_RndM.to_sql('restaurant_inspections', conn, index=False, if_exists='replace')

#SQL query to standardize ZIP codes and join the tables
join_query = '''
WITH standardized_housing AS (
    SELECT
        TRIM("ZIP CODE") AS ZIP_CODE,
        SUM("SITE UNITS") AS Total_Units, -- Predictor variable
        AVG("TDC") AS Avg_TDC
    FROM
        housing_projects
    GROUP BY
        TRIM("ZIP CODE")
),
standardized_restaurants AS (
    SELECT
        TRIM(facility_zip) AS ZIP_CODE,
        AVG(score) AS Avg_Health_Score, -- Potential predicted variable
        COUNT(facility_id) AS Total_Restaurants
    FROM
        restaurant_inspections
    GROUP BY
        TRIM(facility_zip)
)
SELECT
    h.ZIP_CODE,
    h.Total_Units,
    h.Avg_TDC,
    r.Avg_Health_Score,
    r.Total_Restaurants
FROM
    standardized_housing h
CROSS JOIN
    standardized_restaurants r
ON
    h.ZIP_CODE = r.ZIP_CODE
ORDER BY
    h.ZIP_CODE;
'''

#execute the query and fetch the results
joined_data = pd.read_sql_query(join_query, conn)

#perform Linear Regression Analysis
X = joined_data[['Total_Units']]  # Predictor variable
y = joined_data['Avg_Health_Score']  # Response variable

#add a constant to the predictor variable matrix
X = sm.add_constant(X)

#fit the linear regression model
model = sm.OLS(y, X).fit()

#print the summary of the regression analysis
print(model.summary())

#close the connection to the database
conn.close()
```

####**Formulate** a **Hypothesis** and **Perform Statistical Testing**

#####**Hypothesis:** The number of OTP fraud cases (D) has significantly increased from 2018 to 2022.

#####**Statistical Test**

######Paired t-test to compare OTP fraud cases between 2018 and 2022:

```python
from scipy.stats import ttest_rel

# Perform paired t-test comparing OTP fraud cases between 2018 and 2022 across all states/UTs
t_stat, p_value = ttest_rel(dataframe_C['2018 - D'], dataframe_C['2022 - D'])

# Print the t-statistic and p-value to understand the result of the test
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpretation based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: OTP fraud cases have significantly increased from 2018 to 2022.")
else:
    print("Fail to reject the null hypothesis: No significant change in OTP fraud cases from 2018 to 2022.")
```
