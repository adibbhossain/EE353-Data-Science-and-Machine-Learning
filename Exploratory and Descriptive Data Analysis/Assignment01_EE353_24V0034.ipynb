###### Test **Result** and **Converting** Data Types

```python
#For the Housing Project Dataset, we can see columns LAHD FUNDED, LEVERAGE, and JOBS are contain integer value but the data type is float
#To get the accurate value in calculation we should convert them to datatype int

dataframe_H['LAHD FUNDED'] = dataframe_H['LAHD FUNDED'].astype(int) #to convert the 'LAHD FUNDED' columns datatype float to int
dataframe_H['LEVERAGE'] = dataframe_H['LEVERAGE'].astype(int) #to convert the 'LEVERAGE' columns datatype float to int
dataframe_H['JOBS'] = dataframe_H['JOBS'].fillna(0).astype(int) # Replace NaN values with 0
dataframe_H['JOBS'] = dataframe_H['JOBS'].astype(int) #to convert the 'JOBS' columns datatype float to int
```

##### **Checking missing values** and **fill** them with an imputed value **'NULL'** for String and **'0'** for Int and Float
